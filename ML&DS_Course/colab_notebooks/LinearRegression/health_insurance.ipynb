```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("plano_saude.csv")
```

```python
data
```

```python
X = data.iloc[:, 0].values
X
```

```python
y = data.iloc[:, 1].values
y
```

```python
np.corrcoef(X, y)
```

```python
X = X.reshape(-1,1)
X.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
#b0               b1
model.intercept_, model.coef_
```

```python
predictions = model.predict(X)
predictions
```

```python
X.ravel() # Matrix to vector
```

```python
graph = px.scatter(x=X.ravel(), y=y)
graph.add_scatter(x=X.ravel(), y=predictions)
graph.show()
```
