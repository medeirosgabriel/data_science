```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
x = [20,  27,  21,  37,  46, 53, 55,  47,  52,  32,  39,  41,  39,  48,  48]  
y = [1000,1200,2900,1850,900,950,2000,2100,3000,5900,4100,5100,7000,5000,6500]  
```

```python
graph = px.scatter(x = x, y = y)
graph.show()
```

```python
salary_base = np.array([[20,1000],[27,1200],[21,2900],[37,1850],[46,900],
                        [53,950],[55,2000],[47,2100],[52,3000],[32,5900],
                        [39,4100],[41,5100],[39,7000],[48,5000],[48,6500]])
```

```python
scaler_salary = StandardScaler()
salary_base = scaler_salary.fit_transform(salary_base)
```

```python
model = KMeans(n_clusters=3)
model.fit(salary_base)
```

```python
centers = model.cluster_centers_
centers
```

```python
scaler_salary.inverse_transform(model.cluster_centers_)
```

```python
labels = model.labels_
labels
```

```python
graph1 = px.scatter(x = salary_base[:,0], y = salary_base[:,1], color=labels)
graph2 = px.scatter(x = centers[:,0], y = centers[:,1], size = [12, 12, 12])
graph3 = go.Figure(data = graph1.data + graph2.data)
graph3.show()
```
