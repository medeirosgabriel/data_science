#K-Means

##Load Data

```python
from sklearn.datasets import make_blobs
x_random, y_random = make_blobs(n_samples=200, centers=5)
```

```python
import plotly.express as px
graph = px.scatter(x = x_random[:,0], y = x_random[:,1])
graph.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
model.fit(x_random)
```

```python
centers = model.cluster_centers_
centers
```

```python
labels = model.labels_
labels
```

```python
import plotly.graph_objects as go

graph1 = px.scatter(x = x_random[:,0], y = x_random[:,1], color=labels)
graph2 = go.Figure(data = graph1.data)
graph2.show()
```

```python
graph1 = px.scatter(x = x_random[:,0], y = x_random[:,1], color=labels)
graph2 = px.scatter(x = centers[:,0], y = centers[:,1], size = [5, 5, 5, 5, 5])
graph3 = go.Figure(data = graph1.data + graph2.data)
graph3.show()
```
