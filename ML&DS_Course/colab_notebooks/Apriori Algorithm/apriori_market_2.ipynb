```python
import pandas as pd
from apyori import apriori
data = pd.read_csv("mercado2.csv", header=None)
data
```

```python
l = []
for i in range(data.shape[0]):
  l.append([str(data.values[i, j]) for j in range(data.shape[1])])
l, type(l)
```

```python
rules = apriori(l, min_support = 0.003, min_confidence = 0.2, min_lift = 2)
results = list(rules)
results
```

```python
A = []
B = []
support = []
confidence = []
lift = []
for result in results:
  s = result[1]
  result_rules = result[2]
  for result_rule in result_rules:
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    print(a,"-",b,"-",c,"-",l)
    A += [a]
    B += [b]
    support += [s]
    confidence += [c]
    lift += [l]
```

```python
A, B, support, confidence, lift
```

```python
rules_df = pd.DataFrame({"A": A, "B": B, "support": s, "confidence": confidence, "lift": lift})
rules_df.sort_values(by='lift', ascending=False)
```
