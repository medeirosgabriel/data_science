```python
import pickle

with open('credit.pkl', 'rb') as f:
  X_data_train, X_data_test, y_data_train, y_data_test = pickle.load(f)
```

```python
from sklearn.neural_network import MLPClassifier
classifier = MLPClassifier(max_iter=500, verbose=True, 
                           solver='adam', activation='relu',
                           hidden_layer_sizes=(10,10)) # Two layers with 10 neurons 
classifier.fit(X_data_train, y_data_train)
```

```python
predict = classifier.predict(X_data_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(predict, y_data_test)
```

```python
confusion_matrix(predict, y_data_test)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(classifier)
cm.fit(X_data_train, y_data_train)
cm.score(X_data_test, y_data_test)
```
