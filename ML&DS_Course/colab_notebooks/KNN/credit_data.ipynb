```python
import pickle
with open('credit.pkl', mode = 'rb') as f:
  X_data_train, X_data_test, y_data_train, y_data_test = pickle.load(f)
```

#Model Training

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=20, metric="minkowski")
```

```python
classifier.fit(X_data_train, y_data_train)
```

```python
predict = classifier.predict(X_data_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(predict, y_data_test)
```

```python
confusion_matrix(predict, y_data_test)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(classifier)
cm.fit(X_data_train, y_data_train)
cm.score(X_data_test, y_data_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(predict, y_data_test))
```
