```python
import pickle
with open('census.pkl', mode = 'wb') as f:
  X_data_train, y_data_train, X_data_test, y_data_test = pickle.load(f)
```

#Model Training

```python
from sklearn.naive_bayes import GaussianNB
naive = GaussianNB()
naive.fit(X_data_train, y_data_train)
predict = naive.predict(X_data_test)
predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(predict, y_data_test)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(naive);
cm.fit(X_data_train, y_data_train);
cm.score(X_data_test, y_data_test)
```
