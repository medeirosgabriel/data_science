```python
# The objective is predict the Workday alcohol consumption
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv("Data//student-mat.csv")
df2 = pd.read_csv("Data//student-por.csv")
df1.shape, df2.shape # Rows and Columns
```

```python
df1.head()
```

```python
df2.head()
```

```python
data_frames = [df1, df2]
df_f = pd.concat(data_frames)
```

```python
df_f.shape
```

```python
df_f.info()
```

```python
# Object Data / Not a Integer Data
df_f.select_dtypes(include=['object']).columns
```

```python
df_f.school.replace(['GP', 'MS'], [1, 0], inplace=True)
df_f.sex.replace(['M', 'F'], [1, 0], inplace=True)
df_f.address.replace(['U', 'R'], [1, 0], inplace=True)
df_f.famsize.replace(['GT3', 'LE3'], [1, 0], inplace=True)
df_f.Pstatus.replace(['T', 'A'], [1, 0], inplace=True)
df_f.schoolsup.replace(['yes', 'no'], [1, 0], inplace=True)
df_f.famsup.replace(['yes', 'no'], [1, 0], inplace=True)
df_f.activities.replace(['yes', 'no'], [1, 0], inplace=True)
df_f.nursery.replace(['yes', 'no'], [1, 0], inplace=True)
df_f.higher.replace(['yes', 'no'], [1, 0], inplace=True)
df_f.internet.replace(['yes', 'no'], [1, 0], inplace=True)
df_f.romantic.replace(['yes', 'no'], [1, 0], inplace=True)
```

```python
# Data with more than one possible value:
df_f.select_dtypes(include=['object']).columns
```

```python
df_f = pd.get_dummies(df_f)
df_f.info()
```

```python
df_f.isnull().sum().sort_values(ascending=False).head(df_f.shape[1])
```

```python
df_f.corr()
```

```python
df_f.columns
```

```python
for c in df_f.columns:
    if (df_f.corr()['Dalc'][c] < 0.1):
        df_f.drop([c], axis=1)
```

```python
target = df_f['Dalc']
features = df_f.drop('Dalc', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=5, random_state=0) # ML Algorithm - Decision Tree
tree.fit(x_train, y_train)
```

```python
result = tree.score(x_test, y_test)
print("Accuracy:", result)
```
