```python
# Imports

from matplotlib import pyplot as plt
import math

# Data

first_period = {"4-FMCC1": 9.3, "4-IC": 7.9, "4-LP1": 9.4, "4-P1": 9.4, "4-LPT": 7.3}
second_period = {"4-C1": 8.8, "4-FMCC2": 7.1, "4-LP2": 9.7, "4-P2": 9.2, "4-DEC": 9.0}
third_period = {"4-ALG_L": 9.4, "4-EDA": 8.4, "4-LEDA": 9.2, "4-LOG": 8.0, "4-GRAF": 8.9, "2-PE": 10.0}
fourth_period = {"4-BD1": 9.4, "4-PROB": 8.3, "4-LOAC": 10.0, "4-OAC": 10.0, "4-PLP": 9.8}
fifth_period = {"4-EST": 8.3, "4-REDES": 9.8, "4-SO": 9.1, "4-TEO": 9.7}
periods = [first_period, second_period, third_period, fourth_period, fifth_period]

# Math functions

def period_mean(dict):
    sum = 0
    ch = 0
    for key in dict.keys():
        sum += dict[key] * int(key.split("-")[0])
        ch += int(key.split("-")[0])
    return float(sum)/ch

def periods_mean(means):
    sum = 0
    list = []
    for i in range(len(means)):
        sum += means[i]
        list.append(sum/(i + 1))
    return list

# Data visulization

xs = [i for i in range(1, len(periods) + 1)]
period_m = [period_mean(period) for period in periods]

plt.title("Academic Achievement")
plt.xticks(xs)
plt.plot(xs, period_m, 'g-', label='Period Mean')
plt.plot(xs, periods_mean(period_m), 'r-.', label='Periods Mean')
plt.legend(loc=9)
plt.xlabel("Period")
```

```python
def standard_deviation(period):
    mean = period_mean(period)
    sum = 0
    for key in period.keys():
        sum += (period[key] - mean)**2
    return math.sqrt(sum/len(period))

# Data visulization

standard_d = [standard_deviation(period) for period in periods]

plt.xticks(xs)
plt.xlabel("Period")
plt.ylabel("Standard Deviation")
plt.yticks([min(standard_d), max(standard_d)])
plt.plot(xs, standard_d, 'g-', label='Standard Deviation')
plt.title("Standard Deviation - Academic Achievement")
```
